### This script creates predicted versus observed plots for each of the nematode functional group maps (plus the total map)

```r
# Specify the export of the plots as PDF files
pdf(file='Nematode_PredictedVsObserved.pdf')

par(pty="s",mar = c(3.4, 3.2, 3.2, 3.2),oma=c(0.1,0.1,0.1,0.1))
layout(matrix(c(1,1,2,1,1,3,4,5,6), nrow=3, ncol=3, byrow = TRUE))

paletteForUse = c('#d10000',
 '#ff6622',
 '#ffda21',
 '#33dd00',
 '#1133cc',
 '#220066',
 '#330044')

# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Total Number

# Import the raw data
totalNumber = read.csv('/Users/DevinRouth/Downloads/20180723_Nematode_PredVsObs/20180723_TotalNumb_PredVsObs.csv')

# Split the data into predicted versus observed
observedTotal = totalNumber[,c('Total_Numb')]
predictedTotal = totalNumber[,c('classification')]
dataFrameToPlotTotal = data.frame(predictedTotal,observedTotal)

# Use densCols() output to get density at each point
densityTotal = densCols(observedTotal,predictedTotal)
dataFrameToPlotTotal$dens = col2rgb(densityTotal)[1,] + 1L

# Map densities to colors
colors =  colorRampPalette(paletteForUse)(256)
dataFrameToPlotTotal$colors = colors[dataFrameToPlotTotal$dens]

# Plot it, reordering rows so that densest points are plotted on top
plot(predictedTotal~observedTotal,
     data=dataFrameToPlotTotal[order(-dataFrameToPlotTotal$dens),],
     pch=20,
     col=colors,
     cex=.75,
     xlab='',
     ylab='',
     main='Total Number',
     xlim=c(0,15000),
     ylim=c(0,15000),
     cex.main=1.5,
     cex.axis=1.25,
     xaxt='n',
     yaxt='n'
    )
totalXTicks = seq(0, 15000, by=5000)
totalYTicks = seq(0, 15000, by=5000)
axis(1, at=totalXTicks, labels=formatC(totalXTicks, format="f", big.mark = ",", digits=0))
axis(2, at=totalYTicks, labels=formatC(totalYTicks, format="f", big.mark = ",", digits=0))
abline(0, 1)
abline(lm(predictedTotal~observedTotal),lty=5)
title(xlab='Observed',ylab='Predicted',line=2.2,cex.lab=1.5)
R2Total <- round(1 - (sum((observedTotal-predictedTotal )^2)/sum((observedTotal-mean(observedTotal))^2)),2)
legend(x='bottomright',c(as.expression(bquote(R^{2}==.(R2Total)))), cex = 2, bty='n')


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Bacterivores

# Import the raw data
bacterivores = read.csv('/Users/DevinRouth/Downloads/20180723_Nematode_PredVsObs/20180723_Bacterivores_PredVsObs.csv')

# Split the data into predicted versus observed
observedBact = bacterivores[,c('Bacterivor')]
predictedBact = bacterivores[,c('classification')]
dataFrameToPlotBact = data.frame(predictedBact,observedBact)

# Use densCols() output to get density at each point
densityBact = densCols(observedBact,predictedBact)
dataFrameToPlotBact$dens = col2rgb(densityBact)[1,] + 1L

# Map densities to colors
colors =  colorRampPalette(paletteForUse)(256)
dataFrameToPlotBact$colors = colors[dataFrameToPlotBact$dens]

# Plot it, reordering rows so that densest points are plotted on top
plot(predictedBact~observedBact,
     data=dataFrameToPlotBact[order(-dataFrameToPlotBact$dens),],
     pch=20,
     col=colors,
     cex=.75,
     xlab='',
     ylab='',
     main='Bacterivores',
     xlim=c(0,10000),
     ylim=c(0,10000),
     cex.main=1.5,
     cex.axis=1.25,
     xaxt='n',
     yaxt='n'
    )
bactXTicks = seq(0, 8000, by=4000)
bactYTicks = seq(0, 8000, by=4000)
axis(1, at=bactXTicks, labels=formatC(bactXTicks, format="f", big.mark = ",", digits=0))
axis(2, at=bactYTicks, labels=formatC(bactYTicks, format="f", big.mark = ",", digits=0))
abline(0, 1)
abline(lm(predictedBact~observedBact),lty=5)
title(xlab='Observed',ylab='Predicted',line=2.2,cex.lab=1.5)
R2Bacterivores <- round(1 - (sum((observedBact-predictedBact )^2)/sum((observedBact-mean(observedBact))^2)),2)
legend(x='bottomright',c(as.expression(bquote(R^{2}==.(R2Bacterivores)))), cex = 1, bty='n')


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Fungivores

# Import the raw data
fungivores = read.csv('/Users/DevinRouth/Downloads/20180723_Nematode_PredVsObs/20180723_Fungivores_PredVsObs.csv')

# Split the data into predicted versus observed
observedFungi = fungivores[,c('Fungivores')]
predictedFungi = fungivores[,c('classification')]
dataFrameToPlotFungi = data.frame(predictedFungi,observedFungi)

# Use densCols() output to get density at each point
densityFungi = densCols(observedFungi,predictedFungi)
dataFrameToPlotFungi$dens = col2rgb(densityFungi)[1,] + 1L

# Map densities to colors
colors =  colorRampPalette(paletteForUse)(256)
dataFrameToPlotFungi$colors = colors[dataFrameToPlotFungi$dens]

# Plot it, reordering rows so that densest points are plotted on top
plot(predictedFungi~observedFungi,
     data=dataFrameToPlotFungi[order(-dataFrameToPlotFungi$dens),],
     pch=20,
     col=colors,
     cex=.75,
     xlab='',
     ylab='',
     main='Fungivores',
     xlim=c(0,2000),
     ylim=c(0,2000),
     cex.main=1.5,
     cex.axis=1.25,
     xaxt='n',
     yaxt='n'
    )
fungiXTicks = seq(0, 2000, by=1000)
fungiYTicks = seq(0, 2000, by=1000)
axis(1, at=fungiXTicks, labels=formatC(fungiXTicks, format="f", big.mark = ",", digits=0))
axis(2, at=fungiYTicks, labels=formatC(fungiYTicks, format="f", big.mark = ",", digits=0))
abline(0, 1)
abline(lm(predictedFungi~observedFungi),lty=5)
title(xlab='Observed',ylab='Predicted',line=2.2,cex.lab=1.5)
R2Fungi <- round(1 - (sum((observedFungi-predictedFungi )^2)/sum((observedFungi-mean(observedFungi))^2)),2)
legend(x='bottomright',c(as.expression(bquote(R^{2}==.(R2Fungi)))), cex = 1, bty='n')


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Herbivores

# Import the raw data
herbivores = read.csv('/Users/DevinRouth/Downloads/20180723_Nematode_PredVsObs/20180723_Herbivores_PredVsObs.csv')

# Split the data into predicted versus observed
observedHerb = herbivores[,c('Plant_feed')]
predictedHerb = herbivores[,c('classification')]
dataFrameToPlotHerb = data.frame(predictedHerb,observedHerb)

# Use densCols() output to get density at each point
density = densCols(observedHerb,predictedHerb)
dataFrameToPlotHerb$dens = col2rgb(density)[1,] + 1L

# Map densities to colors
colors =  colorRampPalette(paletteForUse)(256)
dataFrameToPlotHerb$colors = colors[dataFrameToPlotHerb$dens]

# Plot it, reordering rows so that densest points are plotted on top
plot(predictedHerb~observedHerb,
     data=dataFrameToPlotHerb[order(-dataFrameToPlotHerb$dens),],
     pch=20,
     col=colors,
     cex=.75,
     xlab='',
     ylab='',
     main='Herbivores',
     xlim=c(0,5000),
     ylim=c(0,5000),
     cex.main=1.5,
     cex.axis=1.25,
     xaxt='n',
     yaxt='n'
    )
herbXTicks = seq(0, 6000, by=2000)
herbYTicks = seq(0, 6000, by=2000)
axis(1, at=herbXTicks, labels=formatC(herbXTicks, format="f", big.mark = ",", digits=0))
axis(2, at=herbYTicks, labels=formatC(herbYTicks, format="f", big.mark = ",", digits=0))
abline(0, 1)
abline(lm(predictedHerb~observedHerb),lty=5)
title(xlab='Observed',ylab='Predicted',line=2.2,cex.lab=1.5)
R2Herb <- round(1 - (sum((observedHerb-predictedHerb )^2)/sum((observedHerb-mean(observedHerb))^2)),2)
legend(x='bottomright',c(as.expression(bquote(R^{2}==.(R2Herb)))), cex = 1, bty='n')


# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Omnivores

# Import the raw data
omnivores = read.csv('/Users/DevinRouth/Downloads/20180723_Nematode_PredVsObs/20180723_Omnivores_PredVsObs.csv')

# Split the data into predicted versus observed
observedOmni = omnivores[,c('Omnivores')]
predictedOmni = omnivores[,c('classification')]
dataFrameToPlotOmni = data.frame(predictedOmni,observedOmni)

# Use densCols() output to get density at each point
density = densCols(observedOmni,predictedOmni)
dataFrameToPlotOmni$dens = col2rgb(density)[1,] + 1L

# Map densities to colors
colors =  colorRampPalette(paletteForUse)(256)
dataFrameToPlotOmni$colors = colors[dataFrameToPlotOmni$dens]

# Plot it, reordering rows so that densest points are plotted on top
plot(predictedOmni~observedOmni,
     data=dataFrameToPlotOmni[order(-dataFrameToPlotOmni$dens),],
     pch=20,
     col=colors,
     cex=.75,
     xlab='',
     ylab='',
     main='Omnivores',
     xlim=c(0,1500),
     ylim=c(0,1500),
     cex.main=1.5,
     cex.axis=1.25,
     xaxt='n',
     yaxt='n'
    )
omniXTicks = seq(0, 2000, by=500)
omniYTicks = seq(0, 2000, by=500)
axis(1, at=omniXTicks, labels=formatC(omniXTicks, format="f", big.mark = ",", digits=0))
axis(2, at=omniYTicks, labels=formatC(omniYTicks, format="f", big.mark = ",", digits=0))
abline(0, 1)
abline(lm(predictedOmni~observedOmni),lty=5)
title(xlab='Observed',ylab='Predicted',line=2.2,cex.lab=1.5)
R2Omni <- round(1 - (sum((observedOmni-predictedOmni )^2)/sum((observedOmni-mean(observedOmni))^2)),2)
legend(x='bottomright',c(as.expression(bquote(R^{2}==.(R2Omni)))), cex = 1, bty='n')

# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Predators

# Import the raw data
predators = read.csv('/Users/DevinRouth/Downloads/20180723_Nematode_PredVsObs/20180723_Predators_PredVsObs.csv')

# Split the data into predicted versus observed
observedPred = predators[,c('Predators')]
predictedPred = predators[,c('classification')]
dataFrameToPlotPred = data.frame(predictedPred,observedPred)

# Use densCols() output to get density at each point
density = densCols(observedPred,predictedPred)
dataFrameToPlotPred$dens = col2rgb(density)[1,] + 1L

# Map densities to colors
colors =  colorRampPalette(paletteForUse)(256)
dataFrameToPlotPred$colors = colors[dataFrameToPlotPred$dens]

# Plot it, reordering rows so that densest points are plotted on top
plot(predictedPred~observedPred,
     data=dataFrameToPlotPred[order(-dataFrameToPlotPred$dens),],
     pch=20,
     col=colors,
     cex=.75,
     xlab='',
     ylab='',
     main='Predators',
     xlim=c(0,1500),
     ylim=c(0,1500),
     cex.main=1.5,
     cex.axis=1.25,
     xaxt='n',
     yaxt='n'
    )
predXTicks = seq(0, 1500, by=500)
predYTicks = seq(0, 1500, by=500)
axis(1, at=predXTicks, labels=formatC(predXTicks, format="f", big.mark = ",", digits=0))
axis(2, at=predYTicks, labels=formatC(predYTicks, format="f", big.mark = ",", digits=0))
abline(0, 1)
abline(lm(predictedPred~observedPred),lty=5)
title(xlab='Observed',ylab='Predicted',line=2.2,cex.lab=1.5)
R2Pred <- round(1 - (sum((observedPred-predictedPred )^2)/sum((observedPred-mean(observedPred))^2)),2)
legend(x='bottomright',c(as.expression(bquote(R^{2}==.(R2Pred)))), cex = 1, bty='n')

dev.off()
```
