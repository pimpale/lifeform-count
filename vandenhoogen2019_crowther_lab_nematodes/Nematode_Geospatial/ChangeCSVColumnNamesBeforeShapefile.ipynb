# This notebook simply changes the column names of the sampled pixel values to reference keys (corresponding to a JSON dictionary) so as to allow upload of the CSV as a Shapefile to Google Earth Engine without violating character limits on column names
## The dictionary to translate the column placeholders to actual covariate names can be found in each of the JS scripts using the Shapefile
## Note: shapefiles were created using QGIS/GDAL

### Import the necessary libraries

```python
import string as string
import itertools as itertools
import pandas as pd

# Define a function to repeat letters of the alphabet like Excel column names
def alphabetSeries(lengthOfSeries):
    def helper():
        n = 1
        while True:
            yield from (''.join(group) for group in itertools.product(string.ascii_uppercase, repeat=n))
            n += 1
    return list(itertools.islice(helper(), lengthOfSeries))
```

```python
# Import the raw data from a CSV file
pathToFile = '20180706_NematodePoints_SampledPixelValues.csv'
rawData = pd.read_csv(pathToFile,float_precision='round_trip')

# Optionally drop columns of disinterest
dataOI = rawData.drop(columns=['system:index','.geo'])

print(dataOI.shape)
print(dataOI.columns)
print(dataOI.head(5))
```

```python
# Create a new list of column names
newColNames = alphabetSeries(len(list(dataOI.columns)))
```

```python
# Make an exportable dictionary of band/column names
keys = list(dataOI.columns)
values = newColNames
bandColNameDictionary = dict(zip(keys, values))
print(bandColNameDictionary)
```

```python
# Rename the columns with placeholders
renamedData = dataOI.rename(index=str, columns=bandColNameDictionary)

print(dataOI.shape)
print(dataOI.columns)
print(dataOI.head(5))

print(renamedData.shape)
print(renamedData.columns)
print(renamedData.head(5))
```

```python
# Export the CSV to convert into a ShapeFile
renamedData.to_csv('20180706_Nematode_SampledData_ColNamePlaceHold.csv',index=False)
```
