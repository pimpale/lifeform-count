<h1 align='center'>This notebook takes the raw (cleaned) data from the nematode dataset and averages all of the values for each unique latitude and longitude value (i.e., for each of the unique ≈1-km pixels represented by the points)<h1>

```python
# Import the necessary packages
import pandas as pd
pd.set_option("display.precision", 8)
```

```python
# Import the raw data from a CSV file
pathToFile = '20180706_Nematode_Points_Pixel_Locations.csv'
rawData = pd.read_csv(pathToFile,float_precision='round_trip')

dataOI = rawData.drop(columns=['system:index','.geo'])

print(dataOI.shape)
print(dataOI.columns)
print(dataOI.head(5))
```

```python
# Remove potential erroneous outliers (i.e., observations with ≥50000 nematodes)
filteredData = dataOI.loc[dataOI['Total_Numb'] < 50000]

print(filteredData.shape)
print(filteredData.columns)
print(filteredData.head(5))
```

## Group the dataset by Lat/Long and take the mean

```python
aggData = pd.DataFrame(filteredData.groupby(['Pixel_Lat', 'Pixel_Long']).mean().to_records())

print(aggData.shape)
print(aggData.columns)
print(aggData.head(5))
```

```python
# Export the result
aggData.to_csv('20180706_Nematode_Data_Aggregated_by_PixelLocation.csv',index=False)
```
