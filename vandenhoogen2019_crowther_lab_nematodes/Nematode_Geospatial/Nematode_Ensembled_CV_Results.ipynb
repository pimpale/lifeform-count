# This notebook is used to cross validate the final ensembled models/maps, producing R<sup>2</sup> scores for each of the functional group models/maps

### Load the necessary libraries

```python
import os as os
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
```

### Total Number

```python
# Load the corresponding functional group CV results
singleFilePath = '/Users/DevinRouth/Downloads/20180720_Nematode_CV_Ensemble_Results/20180720_TotalNumb_AllVars_CV_Ensemble_500Trees/20180720_TotalNumb_AllVars_CV_Ensemble_500Trees.csv'
totalResultsInOneFile = pd.read_csv(singleFilePath,encoding='latin-1',float_precision='round_trip')

# Input the names of the columns used
observedPoI = "Total_Numb"
predictedPoI = "sum"

# Compute the accuracy statistics per fold
grouped = totalResultsInOneFile.groupby('Fold_Number').apply(lambda row: r2_score(row[observedPoI],row[predictedPoI]))
print('Raw values')
print(grouped)
print('Mean R^2')
print(grouped.mean())
print('Standard Deviation of R^2')
print(np.std(grouped))
```
