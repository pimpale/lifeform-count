# This notebook contains code to produce coefficient of determination R<sup>2</sup> values for the grid searched, cross validated model results

### Load the Necessary Libraries

```python
import os as os
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
```

```python
# Input the directory of interest containing all outputted predicted/observed files
directoryOfInterest = '/Users/DevinRouth/Downloads/20180720_Nematode_GridSearch_FullModelRun'
```

```python
# Input the names of the columns used
observedPoI = "Total_Numb"
predictedPoI = "Total_Numb_Predicted"
```

```python
# Retrieve the file names from the directory
fileListInDirectory = os.listdir(directoryOfInterest)
```

```python
# Retrieve the full path of each file in the directory
fullListWithPaths = [directoryOfInterest + '/' + s for s in fileListInDirectory]
# print(fullListWithPaths)

# Make new empty lists for storing the mean R^2 and the standard deviations of the R^2 values
meanValues = list()
stdValues = list()
coefOfVarValues = list()
```

```python
# Run a for loop to compute the statistics for each model results file
for file in fullListWithPaths:
    # Load each of the results list
    resultsDataFrame = pd.read_csv(file,encoding='latin-1',float_precision='round_trip')

    # Compute the accuracy statistics per fold
    grouped = resultsDataFrame.groupby('Fold_Number').apply(lambda row: r2_score(row[observedPoI],row[predictedPoI]))
    
    # Add the values to the empty lists
    meanValues.append(grouped.mean())
    stdValues.append(np.std(grouped))
    coefOfVarValues.append(np.std(grouped)/grouped.mean())
```

```python
# Group all of the values in a single data frame
modelValues = pd.DataFrame(
    {'Model_Title': fileListInDirectory,
     'Mean_Values': meanValues,
     'STD_Values': stdValues,
     'CoefVar_Values': coefOfVarValues
    })
```

```python
# Display each of the model results
pd.set_option('display.max_colwidth', -1)
pd.options.display.max_rows = 999
modelValues.sort_values(by=['Mean_Values'],ascending=False)
```
